# After looking through the data, I've decided I want to find out who were the best catchers to play the game. To start, I'm going to explore the best catchers by looking into the Hall of Fame data.

### Reading & looking at the HOF data

```python
hof = pd.read_csv("/Users/marcmaybaum/Desktop/Education/Udacity/Data_Analyst_Nanodegree/Project_2/Baseball_Stats/HallOfFame.csv")

# print(hof)
```

### Adding position to HOF data so I can find the best catchers

```python
fielding_pos = fielding[['playerID', 'POS']]

hof = hof.merge(fielding_pos, on='playerID', how='left', left_index = True)
```

### Creating a segment of HOF data only for Catchers inducted to the HOF as a player

```python
hof_catcher = (hof[(hof['POS'] == 'C') & (hof['inducted'] == 'Y') & (hof['category'] =='Player')])

hof_catcher = hof_catcher.drop_duplicates()

hof_catcher.shape
```

### Find out these players names by merging with master

```python
hof_catcher = hof_catcher.merge(master, on = 'playerID', how='left', left_index = True)

# print(hof_catcher)
```

# Now, that I've identified Catchers in HOF, I want to learn what their career numbers were for both offense & defense (data found in batting & fielding files), and learn how their production compares to all catchers to understand why they were voted into the HOF.

```python
# Take playerIDs from hof_catcher & find them in fielding & batting files

hof_catcher_playerID = hof_catcher['playerID']

offense_hof_c = batting[batting.playerID.isin(hof_catcher_playerID)]

defense_hof_c = fielding[fielding.playerID.isin(hof_catcher_playerID)]

offense_defense_hof_c = offense_hof_c.merge(defense_hof_c, on=['playerID', 'yearID', 'teamID'], how='inner', left_index=False, right_index=False, suffixes=('_batting','_fielding'))

offense_defense_hof_c.to_csv("/Users/marcmaybaum/Desktop/Education/Udacity/Data_Analyst_Nanodegree/Project_2/Baseball_Stats/OffenseDefense_HOF_C.csv")
```

### Only defensive data I want is for when the player was a catcher. 
### Defensive data includes data for every position player played during the regular season.

```python
offense_defense_hof_c = offense_defense_hof_c[(offense_defense_hof_c['POS'] == 'C')]
```

### After finding the playerIDs from the hof_catcher file and selecting only the defensive stats, I added first and last names to each playerID, formatted the data type of each column to be consistent (float64 for numbers), and replaced NaN with 0 values where applicable.

```python
# Merge these files on playerID, yearID & teamID to have a master catcher offense/defense file

offense_defense_hof_c = offense_defense_hof_c.merge(master, on = 'playerID', how='left', left_index = True)

# Formatting all data types in dataframe

offense_defense_hof_c.dtypes
offense_defense_hof_c = offense_defense_hof_c.astype('float', raise_on_error = False)

# Filling all NaN with 0 so I can create measurements

offense_defense_hof_c = offense_defense_hof_c.fillna(0)

offense_defense_hof_c.to_csv("/Users/marcmaybaum/Desktop/Education/Udacity/Data_Analyst_Nanodegree/Project_2/Baseball_Stats/OffenseDefense_C.csv")
```

## After completing data cleansing steps above, I can find out career averages for each of these stats, when appropriate. 

```python
# First step is looking into the data in the ReadME file & researching baseball stats
# https://en.wikipedia.org/wiki/Baseball_statistics#Batting_statistics

playerID_od_c = offense_defense_hof_c.groupby(['playerID','nameFirst', 'nameLast'])

playerID_od_c = playerID_od_c[['G_batting', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'SB_batting', 'CS_batting', 'BB', 'SO', \
                           'IBB', 'HBP', 'SH', 'SF', 'GIDP', 'G_fielding', 'GS', 'InnOuts', 'PO', 'A', 'E', 'DP', 'PB', \
                           'WP', 'SB_fielding', 'CS_fielding', 'ZR']].sum()
```

### After aggregating the data by playerID, first name, and last name, I am adding some common stats not available in the dataset to analyze players

```python
# Batting average formula based on Wikipedia formula 
# https://en.wikipedia.org/wiki/Baseball_statistics#Batting_statistics

playerID_od_c['AVG'] = (playerID_od_c['H'])/(playerID_od_c['AB'])

# OBP formula retrieved from
# http://www.baseball-reference.com/bullpen/On_base_percentage

playerID_od_c['OBP'] = (playerID_od_c['BB'] + playerID_od_c['H'] + playerID_od_c['HBP'])/(playerID_od_c['AB'] + playerID_od_c['BB'] \
                                                                                 + playerID_od_c['HBP'] + playerID_od_c['SF'])

# SLG Percentage to weight a hitter's impact to runs scored
# https://en.wikipedia.org/wiki/Slugging_percentage

playerID_od_c['SLG'] = ((playerID_od_c['HR'] * 4) + (playerID_od_c['3B'] * 3) +\
                        (playerID_od_c['2B'] * 2) + \
                        (playerID_od_c['H']-(playerID_od_c['2B'] + playerID_od_c['3B'] + playerID_od_c['HR'])))/\
                        playerID_od_c['AB']

# OPS to find total value of player (OBP + SLG)
# http://m.mlb.com/glossary/standard-stats/on-base-plus-slugging 

playerID_od_c['OPS'] = playerID_od_c['SLG'] + playerID_od_c['OBP']

# Writing this to file

playerID_od_c.to_csv("/Users/marcmaybaum/Desktop/Education/Udacity/Data_Analyst_Nanodegree/Project_2/Baseball_Stats/OffenseDefenseAgg_C.csv")

print(playerID_od_c)
```
